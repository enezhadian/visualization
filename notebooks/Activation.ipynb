```python
from os import path
from scipy.ndimage import imread
from IPython.display import display

from visualization import Context
from visualization.visualizers.activation import ActivationVisualizer
from visualization.visualizers import as_image
```

```python
context = Context(path.join('..', 'models', 'tensorflow_inception_graph.pb'))
```

```python
input_image_path = path.join('..', 'resources', 'eagle.jpg')
input_image = imread(input_image_path)
display(as_image(input_image))
```

```python
tensors = [
    ('maxpool0:0', 20),
    ('mixed3a:0', 188),
    ('mixed4e:0', 3)
]

activation_visualizer = ActivationVisualizer(context)
     
for t, i in tensors:
    activation = activation_visualizer.get_activation(tensor=t, index=i, input_image=input_image)
    display(as_image(activation, size=[224, 224]))
```
