```python
from os import path
from IPython.display import display

from visualization import Context
from visualization.visualizers.input import YosinskiEtAlVisualizer
from visualization.visualizers import as_image
```

```python
context = Context(path.join('..', 'models', 'tensorflow_inception_graph.pb'))
```

```python
tensor = 'softmax2_pre_activation:0'
indexes = [630, 397, 104]

yosinskietal_visualizer = YosinskiEtAlVisualizer(context, data_mean=200, gaussian_blur_sigma=5e-1, clip_threshold=5e-2)
     
for i in indexes:
    reconstruction = yosinskietal_visualizer.reconstruct_input(
        tensor_name=tensor,
        unit_index=[0, i],
        iterations=200,
        step=200.0,
        log_per=50
    )
    display(as_image(reconstruction, size=[224, 224]))
```
