```python
tensor = 'softmax2_pre_activation:0'
indexes = [630, 397, 104]

erhanetal_visualizer = ErhanEtAlVisualizer(context, data_mean=200)
     
for i in indexes:
    reconstruction = erhanetal_visualizer.reconstruct_input(
        tensor_name=tensor,
        unit_index=[0, i],
        iterations=100,
        step=100.0,
        log_per=25
    )
    display(as_image(reconstruction, size=[224, 224]))
```
